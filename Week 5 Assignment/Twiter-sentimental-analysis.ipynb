```python
import pandas as pd
```

```python
from textblob import TextBlob

def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    return sentiment
```

```python
df = pd.read_csv('training.1600000.processed.noemoticon.csv',sep=",",encoding = "ISO-8859-1",names=["target", "ids", "date", "flag","user","text"])
```

```python
df.head(100)
```

```python
df = df.assign(sentimental_score = get_sentiment(str(df['text'])))
```
